## Ejemplos de Electricista

```python
# Lista para guardar los nombres de los electricistas
electricistas = []

def agregar_electricista():
    nombre = input("Nombre del electricista: ").strip()
    if nombre:
        electricistas.append(nombre)
        print(f"⚡ {nombre} ha sido dado de alta en el taller.")
    else:
        print("⚠️ El nombre no puede estar vacío.")

def listar_electricistas():
    if not electricistas:
        print("📭 El taller está vacío, no hay electricistas registrados.")
    else:
        print("\n--- LISTA DE ELECTRICISTAS EN TURNO ---")
        for i, nombre in enumerate(electricistas, 1):
            print(f"{i}. {nombre}")

def buscar_electricista():
    nombre_buscar = input("Nombre del electricista a buscar: ").strip()
    if nombre_buscar in electricistas:
        print(f"✅ Confirmado: {nombre_buscar} se encuentra en las instalaciones.")
    else:
        print(f"❌ No se encontró a ningún electricista llamado '{nombre_buscar}'.")

def modificar_electricista():
    nombre_buscar = input("Nombre del electricista a cambiar: ").strip()

    if nombre_buscar in electricistas:
        # Buscamos la posición en la lista para reemplazarlo
        indice = electricistas.index(nombre_buscar)
        nuevo_nombre = input("Ingrese el nuevo nombre: ").strip()
        electricistas[indice] = nuevo_nombre
        print(f"💡 Cambio realizado: '{nombre_buscar}' ahora es '{nuevo_nombre}'.")
    else:
        print("❌ Ese nombre no figura en nuestros registros.")

# --- LÓGICA CON CICLO FOR ---
# El ciclo se ejecutará hasta 50 veces a menos que se use la opción 'Salir'
for intento in range(50):
    print("\n--- GESTIÓN DE PERSONAL (ELECTRICISTAS) ---")
    print("1. Registrar")
    print("2. Ver todos")
    print("3. Localizar")
    print("4. Editar nombre")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_electricista()
    elif opcion == "2":
        listar_electricistas()
    elif opcion == "3":
        buscar_electricista()
    elif opcion == "4":
        modificar_electricista()
    elif opcion == "5":
        print("Saliendo del sistema... ¡Ten un buen día!")
        break
    else:
        print("⚠️ Opción no válida. Intente de nuevo.")

print("\nPrograma realizado por Cepeda Darrell NL = 10")
```
